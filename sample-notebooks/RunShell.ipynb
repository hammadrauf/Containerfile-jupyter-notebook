```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D

rho=28.0
sigma=10.0
beta=8.0/3.0

def f(state,t):
    x, y, z = state
    return sigma*(y-x), x*(rho-z)-y, x*y-beta*z
state0 = [1.0, 1.0, 1.0]
t= np.arange(0.0, 40.0, 0.001)
states= odeint(f,state0,t)

fig=plt.figure(figsize=(16,16),dpi=250)
#ax=fig.gca(projection="3d")
ax = fig.add_subplot(projection = '3d')
ax.plot(states[:,0], states[:,1], states[:,2])
plt.draw()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D

rho=28.0
sigma=10.0
beta=8.0/3.0

def f(state,t):
    x, y, z = state
    return sigma*(y-x), x*(rho-z)-y, x*y-beta*z
state0 = [1.0, 1.0, 1.0]
t= np.arange(0.0, 10.0, 0.001)
states= odeint(f,state0,t)

fig=plt.figure(figsize=(16,16),dpi=250)
#ax=fig.gca(projection="3d")
ax = fig.add_subplot(projection = '3d')
ax.plot(states[:,0], states[:,1], states[:,2])

state0 = [1.01, 1.01, 1.01]
t= np.arange(0.0, 10.0, 0.001)
states= odeint(f,state0,t)
ax.plot(states[:,0], states[:,1], states[:,2],'g')

plt.draw()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
rs=np.linspace(1,4,1000)
N=500
x=0.5+np.zeros(N)
endcap=np.arange(round(N*0.9),N)
f=plt.figure(figsize=(8,8))
for ri in range(len(rs)):
    for n in range(N-1):
        x[n+1]=rs[ri]*x[n]*(1-x[n])
    
    u=np.unique(x[endcap])
    r=rs[ri]*np.ones(len(u))
    plt.plot(r,u,'k.',markersize=1)


plt.show()
```
